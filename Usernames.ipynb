```python
import urllib.request
import re
```

```python
url = "https://filip.journet.sdu.dk/twitter/politikere/"
response = urllib.request.urlopen(url)
data = response.read()      # a `bytes` object
text = data.decode('utf-8')
```

```python
usernames=set(re.findall(r'\@\w+',text))
```
